### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

# Define the geolocation of "Penang Flavours" (latitude and longitude)
penang_flavours_location = {"latitude": 51.49014200, "longitude": 0.08384000}

# Define the aggregation pipeline to find the top 5 establishments
pipeline = [
    {
        "$match": {
            "RatingValue": 5,
            "scores.Hygiene": {"$exists": True},  # Ensure Hygiene score exists
            "geocode": {
                "$geoWithin": {
                    "$box": [
                        [
                            penang_flavours_location["longitude"] - 0.01,
                            penang_flavours_location["latitude"] - 0.01
                        ],
                        [
                            penang_flavours_location["longitude"] + 0.01,
                            penang_flavours_location["latitude"] + 0.01
                        ]
                    ]
                }
            }
        }
    },
    {
        "$sort": {"scores.Hygiene": 1}
    },
    {
        "$limit": 5
    }
]

# Use the aggregate method with the pipeline to fetch the top 5 establishments
results = list(collection.aggregate(pipeline))

# Iterate through the results and print the establishments
for result in results:
    print(result)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Define the aggregation pipeline
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest
pipeline = [
    {
        "$match": {
            "scores.Hygiene": 0  # Match establishments with a hygiene score of 0
        }
    },
    {
        "$group": {
            "_id": "$LocalAuthorityName",  # Group by Local Authority
            "count": {"$sum": 1}  # Count the number of establishments in each group
        }
    },
    {
        "$sort": {"count": -1}  # Sort by count in descending order
    }
]

# Execute the aggregation pipeline
results = list(collection.aggregate(pipeline))

# Print the results
for result in results:
    print("Local Authority:", result["_id"])
    print("Number of establishments with hygiene score 0:", result["count"])
    print()
```
